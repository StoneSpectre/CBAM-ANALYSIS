#### Create Emissions

```python
fuel_map = {
1101001:"Anthracite",
1101002:"Coal",
1101003:"Coal",
1101004:"Coal",
1101006:"Coal",
1101007:"Coal",
1102001:"Coal",
1104000:"Lignite",
1202001:"Natural Gas",
1202003:"LNG",
1202004:"LNG",
1202099:"Natural Gas"
}

cbam_df["fuel_type"] = cbam_df["I13"].map(fuel_map)
```

#### Emission Factors

```python
ef = {
1101001:2.6,
1101002:2.42,
1101003:2.42,
1101004:2.42,
1101006:2.42,
1101007:2.42,
1102001:2.42,
1104000:2.1,
1202001:0.002,
1202003:0.002,
1202004:0.002,
1202099:0.002
}

cbam_df["ef"] = cbam_df["I13"].map(ef)

cbam_df["co2_proxy"] = cbam_df["I15"] * cbam_df["ef"]
```

#### Aggregate Emissions

```python
emissions = (
    cbam_df.groupby("AI01")["co2_proxy"]
    .sum()
    .reset_index()
)
print(emissions.head())
```

#### Merge Emissions Into Firm Dataset

```python
firm_df = firm_df.merge(
    emissions,
    on="AI01",
    how="left"
)
```

#### Compute Carbon Intensity

```python
firm_df["carbon_intensity"] = (
    firm_df["co2_proxy"] /
    firm_df["output_value"]
)
```
