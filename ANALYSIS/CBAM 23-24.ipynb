### Filter CBAM Industries

```python
cbam_codes = [
    23941,23942,
    24101,24102,24103,24104,24105,
    24106,24107,
    24201,24202,24203,24204,
    24209,
    20121,20122,20129,20130
]

cbam_A = A[
    A["a5"].isin(cbam_codes)
]

print(cbam_A.shape)
print(cbam_A["a5"].value_counts())
```

### Keep Exporters Only

```python
cbam_exporters = cbam_A[
    cbam_A["expshare"] > 0
]

print(cbam_exporters.shape)
```

### Employment Block

```python
emp = (
    E.groupby("AE01")
    .sum(numeric_only=True)
    .reset_index()
)

print(emp.head())
```

### Output Block

```python
output = (
    J.groupby("AJ01")["J113"]
    .sum()
    .reset_index()
)

output.columns = [
    "factory",
    "output_value"
]

print(output.head())
```

### Merge Everything

```python
master = cbam_exporters.merge(
    emp,
    left_on="a1",
    right_on="AE01",
    how="left"
)

master = master.merge(
    output,
    left_on="a1",
    right_on="factory",
    how="left"
)

print(master.shape)
print(master.head())
```

### Save Excel
